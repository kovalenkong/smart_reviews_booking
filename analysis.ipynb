Исходная таблица состоит из **11 столбцов**:
- *id*
- *hotel* - название отеля
- *date* - дата отзыва (примечание: дата отзыва представлена в формате ДД.ММ.ГГГГ, однако следует брать во внимание лишь месяц и год - на сайте booking дата отзыва представлена без дня)
- *title* - название отзыва
- *country* - страна, гражданин которой оставил отзыв
- *positive_review* - положиетельный отзыв (если имеется)
- *negative_review* - отрицательный отзыв (если имеется)
- *tr_positive_review* - переведенный положительный отзыв (положительный отзыв переводился, в случае если а) исходный отзыв представлен не на русском или английском языках, если общее количество отзывов данного языка превышало 25)
- *lang_code*: языковой код текстового отзыва

Общая таблица представлена в виде базы данных SQLite (файл reviews.db) и в Excel формате (файл reviews.xlsx).

Также присутствует доп. файл smart_reviews.xlsx, где на двух листах представлены "умные" положительные и отрицательные отзывы.

```python
import numpy as np
import pandas as pd
import sqlite3
import warnings


warnings.filterwarnings('ignore')
```

```python
conn = sqlite3.connect('reviews.db')
df = pd.read_sql('SELECT * FROM reviews;', conn)
```

### Количество оценок по отелям

```python
(df.groupby('hotel')['score']
 .count().to_frame()
 .rename(columns={'score': 'count'})
 .sort_values('count', ascending=False))
```

### Количество оценок по отелям и странам

```python
(df.groupby(['country', 'hotel']).count()['score']
 .unstack().fillna(0).astype(int)
 .style.background_gradient('Blues', axis=0))
```

### Средняя оценка отзвывов и количество отзывов по отелям

```python
(df.groupby('hotel').agg({'score': ['mean', 'count']})
 .sort_values(('score', 'mean'), ascending=False).round({('score', 'mean'): 1}))
```

### Количество положительных и отрицательных отзывов, в которых встречаются такие слова, как: умный/алиса/smart/голосовой

Регулярное выражение: `r'\b(умн[аояе]+|алис[ае]|smart|голосов[ойыеая]+)\s'`

```python
regex = r'\b(умн[аояе]+|алис[ае]|smart|голосов[ойыеая]+)\s'
df[
    (df['positive_review'].notna() & df['positive_review'].str.contains(regex, case=False)) |
    (df['tr_positive_review'].notna() & df['tr_positive_review'].str.contains(regex, case=False))
].shape[0]
```

Всего в 41 положительном отзыве упомянаются "умные" слова.

```python
df[
    (df['positive_review'].notna() & df['positive_review'].str.contains(regex, case=False)) |
    (df['tr_positive_review'].notna() & df['tr_positive_review'].str.contains(regex, case=False))
].groupby('country')['positive_review'].count().to_frame().sort_values('positive_review', ascending=False)
```

Больше всего положительных "умных" отзывов оставили граждане Великобритании и России (16 и 12 отзывов соответственно).

```python
total_positive_reviews = (
    df[df['positive_review'].notna()].groupby('country')['positive_review'].count().to_frame()
)

(df[
    (df['positive_review'].notna() & df['positive_review'].str.contains(regex, case=False)) |
    (df['tr_positive_review'].notna() & df['tr_positive_review'].str.contains(regex, case=False))
].groupby('country')['positive_review'].count()
 .to_frame()
 .assign(total=total_positive_reviews['positive_review'])
 .assign(percent=lambda i: (i['positive_review'] / total_positive_reviews['positive_review'] * 100).round(2))
 .sort_values('percent', ascending=False))
```

В таблице выше представлено количество положительных отзывов, где упомянаются "умные" слова, общее количество положительных отзывов, а также процент "умных" отзывов от общего количества положительных отзывов по странам.

100% положительных отзывов, где упоминаются "умные" слова, приходится на страну "Сент-Винент и Гренадины" (1 из 1 отзывов).
Интересно рассмотреть страны, с бОльшим числоим отзывов: Великобритания, Россия и ОАЭ - на людей из данных стран приходится 0,38%, 0,22% и 0,38% "умных" положительных отзывов соответственно.

```python
total_positive_reviews = (
    df[df['positive_review'].notna()].groupby('hotel')['positive_review'].count().to_frame()
)

(df[
    (df['positive_review'].notna() & df['positive_review'].str.contains(regex, case=False)) |
    (df['tr_positive_review'].notna() & df['tr_positive_review'].str.contains(regex, case=False))
].groupby('hotel')['positive_review'].count()
 .to_frame()
 .assign(total=total_positive_reviews['positive_review'])
 .assign(percent=lambda i: (i['positive_review'] / total_positive_reviews['positive_review'] * 100).round(2))
 .sort_values('percent', ascending=False))
```
